```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("data/PTC.csv")
dataset
pd. set_option('display.max_columns', 100) 
pd. set_option('display.max_rows', None)
```

```python
dataset_columns = (dataset.isnull().sum() > len(dataset)*0.05)
dataset_columns.where(dataset_columns==True).dropna()
```

```python
object_columns = [column for column in dataset.columns if dataset[column].dtype == 'O']
```

```python
def convert_nominal(dataset):
    object_columns = [column for column in dataset.columns if dataset[column].dtype == 'O']
    object_columns = object_columns
    dummies = []
    for column in [object_columns]:
        dummies.append(pd.get_dummies(dataset[column]))
        dataset = dataset.drop(column, axis='columns')
    dataset = pd.concat([dataset, *dummies], axis=1)
    return dataset
```

```python
ds = convert_nominal(dataset)
```
